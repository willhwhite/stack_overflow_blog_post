```python
# wrangle calendar data
```

```python
# function to convert price strings to floats
def price_cleanser(col_name):
    """
    price_cleanser converts dataframe values of a specified column from a string to a float and removes any '$'
    
    param field: the column whose values are to be converted
    return: newly updated column
    """
    cleansed_column = col_name.replace('[\$,]', '', regex=True).astype(float)
    return cleansed_column    
```

```python
# price field: remove $ and convert to float   
calendar['price'] = price_cleanser(calendar['price'])

calendar.head()
```

```python
# wrangle listings data
```

```python
# list out listings fields to see which ones we're interested in
listings.info()
```

```python
# clean prices
listings['price'] = price_cleanser(listings['price'])
listings['weekly_price'] = price_cleanser(listings['weekly_price'])
listings['monthly_price'] = price_cleanser(listings['monthly_price'])
listings['security_deposit'] = price_cleanser(listings['security_deposit'])
listings['cleaning_fee'] = price_cleanser(listings['cleaning_fee'])
listings['extra_people'] = price_cleanser(listings['extra_people'])

# calculated fields for question 1
listings['guests_not_included'] = listings['accommodates'] - listings['guests_included']
listings['additional_guests_cost'] = listings['guests_not_included'] * listings['extra_people']
listings['total_price'] = listings['additional_guests_cost'] + listings['price']
listings['price_per_person'] = listings['total_price'] / listings['accommodates']

# calculated fields for question 2
listings['rating_diff'] = listings['review_scores_rating'] - listings['review_scores_rating'].mean()
listings['price_diff'] = listings['price_per_person'] - listings['price_per_person'].mean()
```

```python
# question 1 - when is the best time to travel to Seattle?
```

```python
calendar['month'] = pd.DatetimeIndex(calendar['date']).month
monthly_price = calendar.groupby(['month']).mean()['price']
monthly_price = monthly_price.to_frame().reset_index()

monthly_price
```

```python
plt.bar(monthly_price['month'], monthly_price['price'])
plt.xlabel('')
plt.xticks(monthly_price['month'])
plt.ylabel('Average Price')
plt.title('Average Prices by Month')
plt.show()
```

```python
monthly_available = calendar.groupby(['month', 'available']).size().unstack(fill_value=0)
monthly_available['availability'] = monthly_available['t']/(monthly_available['t'] + monthly_available['f'])    
    
monthly_available
```

```python
plt.bar(monthly_available.index.values, monthly_available['availability'])
plt.xlabel('Month')
plt.xticks(monthly_available.index.values)
plt.ylabel('Average Availability')
plt.title('Average Availability by Month')
plt.show()
```

```python
# question 2 - which neighborhoods are the cheapest to stay in?
```

```python
q1_cols = ['id', 'zipcode', 'neighbourhood_cleansed', 'price', 'accommodates', 'guests_included', 'extra_people', 'guests_not_included', 'additional_guests_cost', 'total_price', 'price_per_person']
q1_df = listings[q1_cols]

q1_df.head()
```

```python
neighbor_price = q1_df.groupby(['neighbourhood_cleansed']).mean()['price']
neighbor_price = neighbor_price.to_frame().reset_index()
neighbor_price = neighbor_price.sort_values(by=['price'])

neighbor_price
```

```python
chart_data = neighbor_price[:10]

plt.bar(chart_data['neighbourhood_cleansed'], chart_data['price'])
plt.xlabel('Neighborhood')
plt.xticks(chart_data['neighbourhood_cleansed'], rotation=45)
plt.ylabel('Price Per Person ($)')
plt.title('Average Price by Neighborhood')
plt.show()
```

```python
# question 3 - is there an ideal neighborhood based on price and rating?
```

```python
# create subset dataframe of listings data, including only the fields we're interested in
q2_cols = ['id', 'neighbourhood_cleansed', 'price_per_person', 'review_scores_rating', 'rating_diff', 'price_diff']
q2_df = listings[q2_cols]

q2_df.head()
```

```python
# check to see which neighborhoods have the top ratings against the mean
rating_diffs = q2_df.groupby(['neighbourhood_cleansed']).mean()['rating_diff']
rating_diffs = rating_diffs.to_frame().reset_index()
rating_diffs = rating_diffs.sort_values(by=['rating_diff'], ascending=False)

rating_diffs[:10]
```

```python
# check to see which neighborhoods have the top prices against the mean
price_diffs = q2_df.groupby(['neighbourhood_cleansed']).mean()['price_diff']
price_diffs = price_diffs.to_frame().reset_index()
price_diffs = price_diffs.sort_values(by=['price_diff'])

price_diffs[:10]
```

```python
# 'Riverview' is in the top 10 of both, so let's plot both charts for a visual representation
```

```python
# plot the rating difference chart, highlighting the 'Riverview' column
rating_chart = rating_diffs[:10]

neighborhoods = rating_chart['neighbourhood_cleansed']
colors = ["grey" if i != "Riverview" else "red" for i in neighborhoods]

plt.bar(rating_chart['neighbourhood_cleansed'], rating_chart['rating_diff'], color=colors)
plt.xlabel('Neighborhood')
plt.xticks(rating_chart['neighbourhood_cleansed'], rotation=90)
plt.ylabel('Rating Diff')
plt.title('Top Ratings vs Average by Neighborhood')
plt.show()
```

```python
# plot the price difference chart, highlighting the 'Riverview' column
price_chart = price_diffs[:10]

neighborhoods = price_chart['neighbourhood_cleansed']
colors = ["grey" if i != "Riverview" else "red" for i in neighborhoods]

plt.bar(price_chart['neighbourhood_cleansed'], price_chart['price_diff'], color=colors)
plt.xlabel('Neighborhood')
plt.xticks(price_chart['neighbourhood_cleansed'], rotation=90)
plt.ylabel('Price Diff')
plt.title('Best Prices vs Average by Neighborhood')
plt.show()
```
